```python
from Bio import Entrez
Entrez.email = "vik@dataquest.io"
```

```python
def download_sequence(id_code):
    handle = Entrez.esearch(db="nucleotide", term=id_code, retmax="1")
    record = Entrez.read(handle)
    handle = Entrez.efetch(db="nucleotide", id=record["IdList"][0], rettype="fasta", retmode="text")
    return handle.read()
```

```python
sequence_data = {}
for sequence in sequences:
    sequence_data[sequence] = {"fasta": download_sequence(sequence)}
```

```python
from Bio import SeqIO
import io

for k,v in sequence_data.items():
    f = io.StringIO(v["fasta"])
    sequence_data[k]["parsed"] = list(SeqIO.parse(f, "fasta"))[0]
```

```python
from Bio import Align
aligner = Align.PairwiseAligner()

score = aligner.score(sequence_data["NC_045512.2"]["parsed"].seq, sequence_data["OM061695.1"]["parsed"].seq)
```

```python
score
```

```python
import numpy as np

comparisons = np.zeros((4,4))

for i in range(0,4):
    for j in range(i+1,4):
        score = aligner.score(sequence_data[sequences[i]]["parsed"].seq, sequence_data[sequences[j]]["parsed"].seq)
        comparisons[i,j] = score
```

```python
seq1 = "ATATGC"
seq2 = "TTACG"

scores = np.zeros((len(seq1) + 1, len(seq2) + 1))
scores[:,0] = [-1 * i for i in range(0, len(seq1)+1)]
scores[0,:] = [-1 * i for i in range(0, len(seq2)+1)]
```

```python
pointers = {}
indel_score = -1
match_score = 1

for i in range(1, len(seq1) * 2):
    row_sel = i
    col_sel = 1
    while row_sel >= 1 and col_sel < len(seq2)+1:
        if row_sel >= len(seq1)+1:
            row_sel -= 1
            col_sel += 1
            continue
        
        diag = scores[row_sel-1,col_sel-1]
        left = scores[row_sel, col_sel-1] - 1
        top = scores[row_sel-1, col_sel] - 1
        
        if seq1[row_sel-1] == seq2[col_sel-1]:
            diag += 1
        else:
            diag -= 1
        
        combined = [diag, left, top]
        max_combined = max(combined)
        ind = [i for i, j in enumerate(combined) if j == max_combined]
        
        pointer_key = "{}_{}".format(row_sel, col_sel)
        pointers[pointer_key] = []
        if 0 in ind:
            pointers[pointer_key].append((row_sel-1,col_sel-1))
        if 1 in ind:
            pointers[pointer_key].append((row_sel, col_sel-1))
        if 2 in ind:
            pointers[pointer_key].append((row_sel-1, col_sel))
            
        scores[row_sel, col_sel] = max_combined
        row_sel -= 1
        col_sel += 1
```
