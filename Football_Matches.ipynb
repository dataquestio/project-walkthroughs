Data pulled from https://fbref.com/en/comps/9/Premier-League-Stats

```python
import requests
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
data = requests.get(standings_url)
data.text
```

imported 'requests' library which pulls all the HTML from a webpage. This is done through 'requests.get' and is displayed in the console here by calling 'data.text'

Created the variable 'standings_url' and linked it to the EFL webpage, and then called the variable 'standings_url' when using request to .get the data.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data.text)
standings_table = soup.select('table.stats_table') [0]
standings_table
```

Imported Beautiful Soup library that 'parses' the html links, basically it helps to edit down the entire webpage HTML into smaller, more specific, managable chunks. 

We fed in the data.text html with everything, and then specified the variable 'standings_table'. This variable links to every 'table' class found within the html, specifically the ones with 'stats_table' elements included. we used [0] to stop searching after it found the first one, and displayed the results by calling 'standings_table'.

```python
links = standings_table.find_all('a')
links = [l.get("href") for l in links] 
links = [l for l in links if '/squads/' in l]
links
```

instead of using the css to find table elements (as above) we this time used the .find_all to get every 'a' element (i.e. HTML links) within the table. This goes through the table and pulls out only the "href" elements so we can grab each teams link. 

Used 'if' statement to filter out only squad data (not top team scorer & goalkeeper) by saying that only include in list if url includes '/squads/'.

List is then called through 'links'

```python
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

Created variable called 'team_urls' and used 'f' to add in the full urls, the start is added and then the {l} list is added at the end.
